```python
import torch, librosa
from fairseq.models.wav2vec import Wav2VecModel
import wget
from pathlib import Path
import os
```

```python
PRETRAINED_DIR = Path('pretrained/')
os.makedirs(PRETRAINED_DIR, exist_ok=True)
wav2vec_url = 'https://dl.fbaipublicfiles.com/fairseq/wav2vec'
```

```python
file_name = 'wav2vec_big_960h.pt'
file_path = PRETRAINED_DIR / file_name
if not file_path.exists():
    wget.download(f'{wav2vec_url}/{file_name}', str(PRETRAINED_DIR))
```

```python
cp = torch.load(file_path)
model = Wav2VecModel.build_model(cp['args'], task=None)
model.load_state_dict(cp['model'])
model.eval()
```
